```python
import numpy as np
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi']= 160
```

```python
first_run = pd.read_csv("Test_Data/08-29-19 experimenting with values.csv")#, index_col=0)
second_run = pd.read_csv("Test_Data/08-29-19 experimenting with values (last test).csv")#, index_col=0)
dfs = [first_run, second_run]
for f in dfs:
    f["Time"] = f["Time"] / 1000
```

```python
first_run["Delta Humidity"] = first_run["Wet Humidity"] - first_run["Dry Humidity"]
first_run["Time Interval"] = np.gradient(first_run["Time"])
first_run["Wet Hum Gradient"] = np.gradient(first_run["Wet Humidity"])/first_run["Time Interval"]
first_run["Dry Hum Gradient"] = np.gradient(first_run["Dry Humidity"])/first_run["Time Interval"]
# first_run.loc[first_run["Hum"] != 0]
first_run.head()
```

```python
second_run["Delta Humidity"] = second_run["Wet Humidity"] - second_run["Dry Humidity"]
second_run["Time Interval"] = np.gradient(second_run["Time"])
second_run["Wet Hum Gradient"] = np.gradient(second_run["Wet Humidity"])/second_run["Time Interval"]
second_run["Dry Hum Gradient"] = np.gradient(second_run["Dry Humidity"])/second_run["Time Interval"]
second_run.head(10)
```

```python
plt.figure().suptitle('Rate of Change in Dry-side Humidity over Big Fan Speed')
plt.ylabel('Delta Humidity %')
plt.xlabel('Time (min)')
```

```python
first_run.set_index("Delta Humidity")["Dry Hum Gradient"].plot(label="First Run Gradient")
second_run.set_index("Delta Humidity")["Dry Hum Gradient"].plot(label="Second Run Gradient")
plt.legend()
plt.show()
```

```python
plt.figure().suptitle('Rate of Change of Wet-side Humidity over Humidifier Power Output')
plt.ylabel('Delta Humidity %')
plt.xlabel('Hum Speed %')
first_run.set_index("Hum")["Wet Hum Gradient"].plot(label="First Run Gradient")
second_run.set_index("Hum")["Wet Hum Gradient"].plot(label="Second Run Gradient")
plt.legend()
plt.show()
```

```python
plt.figure().suptitle('Humidity over Time in Dry Chamber')
plt.ylabel('Humidity %')
plt.xlabel('Time (min)')
first_run.set_index("Time")["Dry Hum Gradient"].plot(label="humidifier at 100%")
second_run.set_index("Time")["Dry Hum Gradient"].plot(label="humidifier at 50%")
plt.legend()
plt.show()
```
