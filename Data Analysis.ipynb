```python
import numpy as np
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi']= 160
```

```python
ten_micron = "Test_Data/10micron/8-21-19 10micron.csv"
fifteen_micron = "Test_Data/15micron/8-21-19 hum100.csv"
no_mesh = "Test_Data/nomesh/8-20-19-nomesh(humidifier on at 50seconds).csv"
fifteen_micron_hum="Test_Data/15micron/8-27-19 hum50.csv"
ten = pd.read_csv(ten_micron)#, index_col=0)
fifteen = pd.read_csv(fifteen_micron)#, index_col=0)
none = pd.read_csv(no_mesh)#, index_col=0)
halfhum_fifteen = pd.read_csv(fifteen_micron_hum)#, index_col=0)
dfs = [ten, fifteen, none, halfhum_fifteen]
```

```python
# Processing 10 micron data
#Time index of when humidifier was activated was 14875
# start_index = pd.Index(df["Time"]).get_loc(14875)
# print(start_index)

ten["Time"] = (ten["Time"] - 14875) / 60000
ten.head(10)
```

```python
# Processing 15 micron data
# For fifteen micron humidifier started at 13984
fifteen["Time"] = (fifteen["Time"] - 13984) / 60000
fifteen.head(10)
```

```python
# Processing no_mesh data
none["Time"] = (none["Time"] - 50000) / 60000
none.head(10)
```

```python
# Processing 15 micron data
#Time index of when humidifier was activated was 2307
halfhum_fifteen["Time"] = (halfhum_fifteen["Time"] - 2307) / 60000
halfhum_fifteen.head(10)
```

```python
for f in dfs:
    f.set_index("Time", inplace=True)
```

```python
plt.figure().suptitle('Humidity over Time in Wet Chamber')
plt.ylabel('Humidity %')
plt.xlabel('Time (min)')
ten["Wet Humidity"].plot(label="ten micron")
fifteen["Wet Humidity"].plot(label="fifteen micron")
none["Wet Humidity"].plot(label="no mesh")
plt.legend()
plt.show()
```

```python
plt.figure().suptitle('Humidity over Time in Dry Chamber')
plt.ylabel('Humidity %')
plt.xlabel('Time (min)')
ten["Dry Humidity"].plot(label="ten micron")
fifteen["Dry Humidity"].plot(label="fifteen micron")
none["Dry Humidity"].plot(label="none")
plt.legend()
plt.show()
```

```python
plt.figure().suptitle('Humidity over Time in Wet Chamber')
plt.ylabel('Humidity %')
plt.xlabel('Time (min)')
fifteen["Wet Humidity"].plot(label="humidifier at 100%")
halfhum_fifteen["Wet Humidity"].plot(label="humidifier at 50%")
plt.legend()
plt.show()
```

```python
plt.figure().suptitle('Humidity over Time in Dry Chamber')
plt.ylabel('Humidity %')
plt.xlabel('Time (min)')
fifteen["Dry Humidity"].plot(label="humidifier at 100%")
halfhum_fifteen["Dry Humidity"].plot(label="humidifier at 50%")
plt.legend()
plt.show()
```
